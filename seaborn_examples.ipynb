```python
# import numpy and seaborn packages
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# set the seaborn style
sns.set(style="ticks")
```

```python
# Mersenne Twister pseudo-random number generator
# random seed = 11
rands = np.random.RandomState(11)
```

```python
# create a gamma distribution for x
# create a normal distribution for y based on x
x = rands.gamma(2, size = 1000)
y = -.2 * x + rands.normal(size=1000)
```

```python
# create a seaborn joint distribution plot
a = sns.jointplot(x, y, kind="reg", color="lightblue")
plt.savefig('figures/joint_distribution.pdf')  
```

```python
# create a seaborn joint distribution plot using hexagons
b = sns.jointplot(x, y, kind="hex", color="lightblue")
plt.savefig('figures/joint_distribution_hex.pdf') 
```

```python
# create a seaborn joint distribution plot using kernel density estimation
c = sns.jointplot(x, y, kind="kde", height=7, space=0, color="lightblue")
plt.savefig('figures/joint_distribution_kde.pdf') 
```

```python
# create a scatterplot of x and y
# show y as the hue and x as the size
cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True)
d = sns.scatterplot(x, y,
                     hue=y, size=x,
                     palette=cmap, sizes=(10, 200))
plt.xlabel('Gamma Distribution')
plt.ylabel('Normal Distribution')
plt.title("Scatter Plot")
plt.savefig('figures/scatterplot.pdf') 
```
